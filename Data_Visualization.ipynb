```python
av = np.array([44.48,72.13,
66.79,45.9,
51.39,81.75,
59.08,70.16])
sr = np.array([89.75,98.46,
87.7,93.29,
94.71,99.39,
90.74,96.84
])
group = ['England',	
'New Zealand',
'South Africa',
'Pakistan',	
'Australia',	
'Bangladesh',
'Sri Lanka',	
'West Indies']

cdict = {'England' : 'red',	
'New Zealand': 'blue',
'South Africa': 'green',
'Pakistan':'yellow',	
'Australia': 'orange',	
'Bangladesh': 'violet',
'Sri Lanka': 'cyan',	
'West Indies': 'black'}

fig, ax = plt.subplots()
for i,g in enumerate(group):
    ax.scatter(av[i], sr[i], c = cdict[g], label = g, s = 100)
ax.legend()
plt.xlabel('average')
plt.ylabel('strike rates')
plt.savefig('avsrteam-scatter.png')
plt.show()
```
